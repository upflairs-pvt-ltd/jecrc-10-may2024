```python
days = list(df['day'].value_counts().keys())
```

```python
valu =  list(df['day'].value_counts().values)
```

```python
plt.bar(days,valu)   # barplot  
plt.ylabel("customer counts")
plt.xlabel('Days name ') 
plt.title("customer counts days wise") 
plt.show()
```

```python
plt.bar(days,valu,color='red')   # barplot  
plt.ylabel("customer counts")
plt.xlabel('Days name ') 
plt.title("customer counts days wise") 
plt.show()
```

```python
plt.bar(days,valu,color='red',ec='b')   # barplot  ec ==> edge color   
plt.ylabel("customer counts")
plt.xlabel('Days name ') 
plt.title("customer counts days wise") 
plt.show()
```

```python
sns.barplot(x=days,y=valu)  # plt.bar() 
```

```python
country_population = [657, 845, 852, 951, 960] 

country_name = ['us','ind','ast','pak','bangladesh']
```

```python
plt.plot(country_name,country_population,marker="*",ms=14)
```

```python
plt.plot(country_name,country_population,marker="*",ms=14,color='red',mfc='blue')
plt.xlabel("a") 
plt.ylabel('b') 
plt.title('lineplot') 
plt.show()
```

```python
plt.plot(country_name,country_population,marker="*",ms=14,color='red',mfc='blue',ls='dotted')
plt.xlabel("a") 
plt.ylabel('b') 
plt.title('lineplot') 
plt.show()
```

```python
plt.plot(country_name,country_population,marker="*",ms=14,color='red',mfc='blue',ls='dashed')
plt.xlabel("a") 
plt.ylabel('b') 
plt.title('lineplot') 
plt.show()
```
